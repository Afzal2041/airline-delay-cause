```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

```

```python
df = pd.read_csv("Airline_Delay_Cause.csv")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    data=df,
    x="carrier_name",
    y="arr_delay",
    hue="carrier_name",
    estimator=sum,
    palette="viridis",
    legend=False
)
plt.xticks(rotation=90)
plt.title("Total Arrival Delay by Airline")
plt.ylabel("Total Delay (minutes)")
plt.xlabel("Airline")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.lineplot(
    data=df,
    x="year",
    y="arr_delay",
    estimator=sum,
    marker="o"
)
plt.title("Total Arrival Delay Over Years")
plt.xlabel("Year")
plt.ylabel("Total Delay (minutes)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["arr_delay"],
    bins=50,
    kde=True,
    color="steelblue"
)
plt.title("Distribution of Arrival Delays")
plt.xlabel("Arrival Delay (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
corr = df.select_dtypes("number").corr()

plt.figure(figsize=(12,8))
sns.heatmap(
    corr,
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation Heatmap of Delay Factors")
plt.show()
```
